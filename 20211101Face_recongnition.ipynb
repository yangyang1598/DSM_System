```python
import face_recognition 

image = face_recognition.load_image_file("MASK.jpg") 

face_locations = face_recognition.face_locations(image)
```

```python
#모델의 결과,(top,right,bottom,left)
face_locations
```

```python
import cv2
import matplotlib.pyplot as plt


for (y1,x2,y2,x1) in face_locations:
  cv2.rectangle(image,(x1,y1),(x2,y2),(0,255,20),3)
plt.imshow(image)
plt.show
```
